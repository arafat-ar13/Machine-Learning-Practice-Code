```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
pokemon_df = pd.read_csv("pokemon_data.csv")
pokemon_df.head()
```

<span style="font-size: 3em;">Selecting specific columns with Pandas is super easy</span>

```python
# Looking at the unique types of Pokemon
pokemon_df["Type 1"].unique()
```

<span style="color: #f2cf4a; font-family: Calibri; font-size: 2em;">We can select various types of Pokemon and save them in a variable for later use</span>

```python
grass_pokemon = pokemon_df[pokemon_df["Type 1"] == "Grass"]
fire_pokemon = pokemon_df[pokemon_df["Type 1"] == "Fire"]
water_pokemon = pokemon_df[pokemon_df["Type 1"] == "Water"]
electric_pokemon = pokemon_df[pokemon_df["Type 1"] == "Electric"]

# Let's view electric_pokemon dataframe for example
electric_pokemon.head(13)
```

<span style="font-family:Calibri; font-size: 2em;">Let's see the number of Pokemon in each of those types</span>

```python
print(grass_pokemon.Name.count())
print(fire_pokemon.Name.count())
print(water_pokemon.Name.count())
print(electric_pokemon.Name.count())
```

## We see a lot of "NaN" values for the "Type 2" of these Pokemon types. Let's view the Fire type Pokemon that also have a "Type 2"

```python
fire_pokemon_with_second_type = fire_pokemon[fire_pokemon["Type 2"].isnull() == False]
fire_pokemon_with_second_type
```

<span style="font-size: 3em;">Learning the Groupby() method</span>

## Groupby does three things: Splits your selected column, Applies any function, Combines them for display
## Let's work with Fire Pokemon with Second Type for simplicity
## Here's, let's use the Groupby method on "Type 2" since "Type 1" for all of them is Fire. Let's view the mean Attack of each type of "Type 2"s 

```python
print()
print("Fire Pokemon 2nd type attack mean:")
type2_attack_grouped = fire_pokemon_with_second_type.groupby("Type 2").Attack.mean().astype(int).reset_index()
type2_attack_grouped
```

<span style="font-size: 1em;">Therefore we learn the mean attack points of each of the "Type 2" Pokemon of the Fire Pokemon</span>

# We can also groupby by two columns for a broader comparison
## Let's groupby by Type 2 like before but now also by Generation. So, now, we can see the Type 2 of each Generation and their mean Attack points

```python
fire_grouped = fire_pokemon_with_second_type.groupby(["Type 2", "Generation"]).Attack.mean().astype(int).reset_index() # We pass in the two columns as a list
fire_grouped
```

# Come the Pivot() method!

```python
fire_grouped_pivoted = fire_grouped.pivot(index="Type 2", columns="Generation", values="Attack")
fire_grouped_pivoted
```

# This view is much preferred than the earlier one as this helps us understand better.
## But we also got a ton of NaNs. We can clean this up using Pandas fillna() method

```python
fire_grouped_pivoted.fillna("No value", inplace=True)
fire_grouped_pivoted
```

<span style="font-size: 3em; color: aqua; font-family: calibri">Plotting with MatPlotLib!!</span>

```python
types = pokemon_df["Type 1"].unique()
types_num = range(len(types))
```

```python
poke_types_count_dict = {key:0 for key in types}

for x in range(800): # 800 is the number of pokemon in our data set
    for type in poke_types_count_dict:
        poke_types_count_dict[type] = pokemon_df[pokemon_df["Type 1"] == type].Name.count()
```

# A pie chart showing the distribution of all the different types of Pokemon

```python
plt.pie(list(poke_types_count_dict.values()), labels=list(poke_types_count_dict), colors=["limegreen", "orangered", "deepskyblue", "seagreen", "wheat", "slateblue", "yellow", "darkgoldenrod", "gold", "firebrick", "darkmagenta", "tan", "gray", "aqua", "red", "midnightblue", "slategrey", "y"], radius=3, autopct='%.2f%%')
plt.show()
```

```python
poke_types_attack_mean = {pokemon: 0 for pokemon in types}

for x in range(800):
    for pokemon in poke_types_attack_mean:
        poke_types_attack_mean[pokemon] = int(pokemon_df[pokemon_df["Type 1"] == pokemon].Attack.mean())
```

```python
plt.figure(figsize=(16, 9))

ax = plt.subplot()

plt.bar(types_num, list(poke_types_attack_mean.values()))
plt.title("The mean of the Attack of each type of Pokemon")
plt.xlabel("Types")
plt.ylabel("Attack Points")

ax.set_xticks(types_num)
ax.set_xticklabels(list(pokemon_df["Type 1"].unique()))

x = plt
```
